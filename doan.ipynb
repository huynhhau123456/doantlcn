```python
merge_df = pd.merge(df_rating, df_movie, on='movieId')
print(merge_df)
```

```python
#có 9724 bộ phim đã được đánh giá
test = merge_df['movieId'].nunique()
print(test)
print(merge_df.dtypes)
```

```python
user_ratings = merge_df[merge_df['userId'] == 2]
abc = user_ratings.iloc[0:10]
xyz = user_ratings[~user_ratings.isin(abc)].dropna()
print(xyz)
```

```python
user_movie_matrix = merge_df.pivot(index='userId', columns='movieId', values='rating')
user_movie_matrix = user_movie_matrix.fillna(0)
print(user_movie_matrix)
```

```python
user_similarity = cosine_similarity(user_movie_matrix)
user_similarity_df = pd.DataFrame(user_similarity, index=user_movie_matrix.index, columns=user_movie_matrix.index)
```

```python
print(user_similarity_df)
```

```python
user_id=2
k=5

# Lấy ra cột tương ứng với user_id
similarity_series = user_similarity_df.loc[user_id]

# Sắp xếp giảm dần và lấy ra k giá trị đầu tiên, bỏ giá trị đầu (iloc[1:]) vì nó trùng với người đang xét
top_k_users = similarity_series.sort_values(ascending=False).head(k + 1).iloc[1:]
```

```python
print("Top", k, "users similar to user", user_id, "are: ")
print(top_k_users)
```

```python
new_user_id = 2

# Lấy ra tất cả người dùng (loại bỏ người dùng mới)
existing_users = user_similarity_df.columns.difference([new_user_id]) #có 609 thằng

# Lấy ra tất cả các bộ phim mà người dùng mới chưa đánh giá
#user_ratings = merge_df[merge_df['userId'] == new_user_id] #có 29 bộ phim đã đc đánh giá
#unrated_movies = merge_df[~merge_df.isin(user_ratings)].dropna() #100836 - 29 = 100807
#unrated_movies['movieId'] = unrated_movies['movieId'].astype(int)
#unrated_movies['userId'] = unrated_movies['userId'].astype(int)
#unrated_movies['timestamp'] = unrated_movies['timestamp'].astype(int)

#những bộ phim mà người dùng chưa xem (chưa đánh giá)
unrated_movies = user_movie_matrix.columns[user_movie_matrix.loc[new_user_id] == 0]
print(unrated_movies)


#unrated_movies = existing_users
#print(user_ratings)
```

```python
predicted_ratings = {}
for movie in unrated_movies:
    # Lọc ra những người dùng đã đánh giá phim này
    users_rated = user_movie_matrix.index[~user_movie_matrix[movie].isna()]
    
    # Tính weighted sum của ratings và cosine similarity
    numerator = (user_movie_matrix.loc[users_rated, movie] * user_similarity_df.loc[users_rated, new_user_id]).sum()
    denominator = user_similarity_df.loc[users_rated, new_user_id].sum()
    
    # Tính và lưu predicted rating
    if denominator != 0:
        predicted_ratings[movie] = numerator / denominator
print(predicted_ratings)
```

```python
# Nếu có ít nhất một đánh giá dự đoán, gợi ý phim có đánh giá cao nhất
if predicted_ratings:
    recommended_movie = max(predicted_ratings, key=predicted_ratings.get)
    print(f"Phim được gợi ý cho người dùng {new_user_id} là movie có thứ tự {recommended_movie} với đánh giá dự đoán là {predicted_ratings[recommended_movie]}")
else:
    print(f"Không có phim nào để gợi ý cho {new_user_id}")
```
